# Pixel Values and Grayscale Conversion

## Overview
This notebook explores pixel representation, grayscale conversion methods, and color space fundamentals. We'll demonstrate:
- RGB to Grayscale conversion using different methods
- Color space transformations (RGB, HSV, YCbCr, LAB)
- Pixel value calculations and formulas
- Visual comparisons of color spaces

## Part 1: RGB to Grayscale Conversion Methods

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load image
img_path = 'sample.jpeg'
image = cv2.imread(img_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Split channels
r, g, b = cv2.split(image_rgb)

print("=" * 60)
print("RGB TO GRAYSCALE CONVERSION METHODS")
print("=" * 60)

# Method 1: Standard formula (ITU-R BT.601)
gray_standard = 0.299 * r + 0.587 * g + 0.114 * b
gray_standard = np.clip(gray_standard, 0, 255).astype(np.uint8)

# Method 2: Simple average
gray_average = (r.astype(np.float32) + g.astype(np.float32) + b.astype(np.float32)) / 3
gray_average = np.clip(gray_average, 0, 255).astype(np.uint8)

# Method 3: OpenCV cvtColor (uses standard formula)
gray_opencv = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Method 4: Luminosity method
gray_luminosity = 0.2126 * r + 0.7152 * g + 0.0722 * b
gray_luminosity = np.clip(gray_luminosity, 0, 255).astype(np.uint8)

# Method 5: Maximum component
gray_max = np.maximum(np.maximum(r, g), b)

# Method 6: Minimum component
gray_min = np.minimum(np.minimum(r, g), b)

print("\n6 Different Grayscale Conversion Methods:")
print("\n1. Standard ITU-R BT.601:  Gray = 0.299×R + 0.587×G + 0.114×B")
print("2. Simple Average:          Gray = (R + G + B) / 3")
print("3. OpenCV cvtColor:         Uses standard formula internally")
print("4. Luminosity (ITU-R Rec.709): Gray = 0.2126×R + 0.7152×G + 0.0722×B")
print("5. Maximum Component:       Gray = max(R, G, B)")
print("6. Minimum Component:       Gray = min(R, G, B)")

# Visualize all methods
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

grayscale_versions = [
    (gray_standard, "Standard ITU-R BT.601"),
    (gray_average, "Simple Average"),
    (gray_opencv, "OpenCV cvtColor"),
    (gray_luminosity, "Luminosity (ITU-R 709)"),
    (gray_max, "Maximum Component"),
    (gray_min, "Minimum Component"),
]

for idx, (gray_img, title) in enumerate(grayscale_versions):
    row = idx // 3
    col = idx % 3
    axes[row, col].imshow(gray_img, cmap='gray')
    axes[row, col].set_title(title)
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```

## Part 2: Pixel-Level Grayscale Conversion Calculations

```python
# Analyze specific pixel conversions
print("\n" + "=" * 60)
print("PIXEL-LEVEL GRAYSCALE CONVERSION EXAMPLES")
print("=" * 60)

# Select various pixels
h, w = image.shape[:2]
test_pixels = [
    (0, 0),
    (0, w-1),
    (h-1, 0),
    (h-1, w-1),
    (h//2, w//2),
    (h//4, w//4),
]

print("\nPixel RGB → Grayscale Conversions:")
print(f"{'Position':<15} {'R':<5} {'G':<5} {'B':<5} {'Standard':<10} {'Average':<10} {'Luminosity':<10} {'Max':<5} {'Min':<5}")
print("-" * 90)

for y, x in test_pixels:
    pixel_rgb = image_rgb[y, x]
    r, g, b = pixel_rgb[0], pixel_rgb[1], pixel_rgb[2]
    
    # Calculations
    gray_std = 0.299*r + 0.587*g + 0.114*b
    gray_avg = (r + g + b) / 3
    gray_lum = 0.2126*r + 0.7152*g + 0.0722*b
    gray_max = max(r, g, b)
    gray_min = min(r, g, b)
    
    pos = f"({y},{x})"
    print(f"{pos:<15} {r:<5} {g:<5} {b:<5} {gray_std:<10.1f} {gray_avg:<10.1f} {gray_lum:<10.1f} {gray_max:<5} {gray_min:<5}")

# Statistical comparison
print("\n" + "=" * 60)
print("STATISTICAL COMPARISON OF METHODS")
print("=" * 60)

methods = [
    (gray_standard, "Standard ITU-R"),
    (gray_average, "Simple Average"),
    (gray_luminosity, "Luminosity"),
    (gray_max, "Max Component"),
    (gray_min, "Min Component"),
]

print(f"\n{'Method':<20} {'Min':<8} {'Max':<8} {'Mean':<8} {'Std':<8}")
print("-" * 50)

for gray_img, method_name in methods:
    print(f"{method_name:<20} {gray_img.min():<8} {gray_img.max():<8} {gray_img.mean():<8.2f} {gray_img.std():<8.2f}")
```

## Part 3: Why Different Weights? Understanding the Formula

```python
# Demonstrate why human eye is more sensitive to green
print("\n" + "=" * 60)
print("WHY DIFFERENT WEIGHTS IN GRAYSCALE FORMULA?")
print("=" * 60)

print("\nHuman Vision Sensitivity:")
print("-" * 40)
print("The human eye is not equally sensitive to all colors.")
print("It's most sensitive to GREEN light, and least to BLUE light.")
print("\nITU-R BT.601 Weights (Standard):")
print("  Red:   0.299 (30%)")
print("  Green: 0.587 (59%)")
print("  Blue:  0.114 (11%)")
print("  Total: 1.000 (100%)")

print("\nComparison of Weight Schemes:")
print(f"{'Method':<20} {'Red':<8} {'Green':<8} {'Blue':<8} {'Total':>8}")
print("-" * 50)
print(f"{'ITU-R BT.601':<20} {'0.299':<8} {'0.587':<8} {'0.114':<8} {'1.000':>8}")
print(f"{'ITU-R Rec.709':<20} {'0.2126':<8} {'0.7152':<8} {'0.0722':<8} {'1.000':>8}")
print(f"{'Simple Average':<20} {'0.333':<8} {'0.333':<8} {'0.333':<8} {'1.000':>8}")

print("\nWhy Green has highest weight?")
print("1. Human eye has more cone cells sensitive to green wavelengths")
print("2. Green light (540-570 nm) is in the peak sensitivity range")
print("3. Red light (630-700 nm) has lower absolute sensitivity")
print("4. Blue light (420-440 nm) has the lowest sensitivity")
print("5. This weighting produces more perceptually accurate grayscale images")
```

```python
# Convert to different color spaces
print("\n" + "=" * 60)
print("COLOR SPACE TRANSFORMATIONS")
print("=" * 60)

# HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h_channel = hsv[:, :, 0] * 2  # OpenCV uses 0-180, multiply for 0-360 degrees
s_channel = hsv[:, :, 1]
v_channel = hsv[:, :, 2]

# YCrCb
ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
y_channel = ycrcb[:, :, 0]
cr_channel = ycrcb[:, :, 1]
cb_channel = ycrcb[:, :, 2]

# LAB
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
l_channel = lab[:, :, 0]
a_channel = lab[:, :, 1]
b_channel = lab[:, :, 2]

# Visualize color spaces
fig, axes = plt.subplots(3, 4, figsize=(16, 12))

# Row 1: RGB
axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title('Original RGB')
axes[0, 0].axis('off')

r, g, b = cv2.split(image_rgb)
axes[0, 1].imshow(r, cmap='Reds')
axes[0, 1].set_title('Red Channel')
axes[0, 1].axis('off')

axes[0, 2].imshow(g, cmap='Greens')
axes[0, 2].set_title('Green Channel')
axes[0, 2].axis('off')

axes[0, 3].imshow(b, cmap='Blues')
axes[0, 3].set_title('Blue Channel')
axes[0, 3].axis('off')

# Row 2: HSV
axes[1, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1, 0].set_title('Original (for reference)')
axes[1, 0].axis('off')

axes[1, 1].imshow(h_channel, cmap='hsv')
axes[1, 1].set_title('Hue (0-360°)')
axes[1, 1].axis('off')

axes[1, 2].imshow(s_channel, cmap='gray')
axes[1, 2].set_title('Saturation (0-255)')
axes[1, 2].axis('off')

axes[1, 3].imshow(v_channel, cmap='gray')
axes[1, 3].set_title('Value/Brightness (0-255)')
axes[1, 3].axis('off')

# Row 3: YCrCb
axes[2, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[2, 0].set_title('Original (for reference)')
axes[2, 0].axis('off')

axes[2, 1].imshow(y_channel, cmap='gray')
axes[2, 1].set_title('Y (Luminance)')
axes[2, 1].axis('off')

axes[2, 2].imshow(cr_channel, cmap='gray')
axes[2, 2].set_title('Cr (Red Chrominance)')
axes[2, 2].axis('off')

axes[2, 3].imshow(cb_channel, cmap='gray')
axes[2, 3].set_title('Cb (Blue Chrominance)')
axes[2, 3].axis('off')

plt.tight_layout()
plt.show()

print("\nColor Space Details:")
print("\nRGB (Red-Green-Blue):")
print(f"  R range: [{r.min()}, {r.max()}]  Mean: {r.mean():.2f}")
print(f"  G range: [{g.min()}, {g.max()}]  Mean: {g.mean():.2f}")
print(f"  B range: [{b.min()}, {b.max()}]  Mean: {b.mean():.2f}")

print("\nHSV (Hue-Saturation-Value):")
print(f"  H range: [0°, 360°] (OpenCV: 0-180)  Mean: {hsv[:,:,0].mean():.2f}")
print(f"  S range: [0, 255]  Mean: {hsv[:,:,1].mean():.2f}")
print(f"  V range: [0, 255]  Mean: {hsv[:,:,2].mean():.2f}")

print("\nYCrCb (Luminance-Chrominance):")
print(f"  Y range:  [0, 255]  Mean: {y_channel.mean():.2f}")
print(f"  Cr range: [0, 255]  Mean: {cr_channel.mean():.2f}")
print(f"  Cb range: [0, 255]  Mean: {cb_channel.mean():.2f}")

print("\nLAB (Perceptually Uniform):")
print(f"  L range:  [0, 255]  Mean: {l_channel.mean():.2f}")
print(f"  A range:  [0, 255]  Mean: {a_channel.mean():.2f}")
print(f"  B range:  [0, 255]  Mean: {b_channel.mean():.2f}")
```

## Part 6: Real-World Pixel Calculation Examples

```python
# Real-world examples with calculations
print("\n" + "=" * 80)
print("REAL-WORLD PIXEL CALCULATION EXAMPLES")
print("=" * 80)

examples = [
    ("Pure Red", 255, 0, 0),
    ("Pure Green", 0, 255, 0),
    ("Pure Blue", 0, 0, 255),
    ("White", 255, 255, 255),
    ("Black", 0, 0, 0),
    ("Yellow (R+G)", 255, 255, 0),
    ("Cyan (G+B)", 0, 255, 255),
    ("Magenta (R+B)", 255, 0, 255),
    ("Medium Gray", 128, 128, 128),
    ("Orange", 255, 165, 0),
]

print(f"\n{'Color':<20} {'R':<8} {'G':<8} {'B':<8} {'Standard':<12} {'Luminosity':<12} {'Average':<12}")
print("-" * 90)

for color_name, r, g, b in examples:
    # Standard formula
    gray_std = 0.299*r + 0.587*g + 0.114*b
    # Luminosity formula
    gray_lum = 0.2126*r + 0.7152*g + 0.0722*b
    # Average
    gray_avg = (r + g + b) / 3
    
    print(f"{color_name:<20} {r:<8} {g:<8} {b:<8} {gray_std:<12.2f} {gray_lum:<12.2f} {gray_avg:<12.2f}")

# Visualization of these colors
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
axes = axes.flatten()

for idx, (color_name, r, g, b) in enumerate(examples[:-1]):  # Exclude last one to fit 9 cells
    # Create a color swatch
    color_rgb = np.ones((100, 100, 3), dtype=np.uint8)
    color_rgb[:, :, 0] = r
    color_rgb[:, :, 1] = g
    color_rgb[:, :, 2] = b
    
    axes[idx].imshow(color_rgb)
    
    gray_std = 0.299*r + 0.587*g + 0.114*b
    axes[idx].set_title(f"{color_name}\nGray: {gray_std:.0f}")
    axes[idx].axis('off')

axes[-1].axis('off')
plt.tight_layout()
plt.show()
```

## Part 7: Illumination-Reflectance Effects on Color

```python
# Demonstrate how illumination affects pixel values
print("\n" + "=" * 80)
print("ILLUMINATION EFFECTS ON PIXEL VALUES")
print("=" * 80)

# Create images with different illumination levels
original = image_rgb.copy()

# Apply different illumination multipliers
illumination_levels = [0.5, 0.75, 1.0, 1.25, 1.5]
illuminated_images = []

for level in illumination_levels:
    illuminated = (original.astype(np.float32) * level)
    illuminated = np.clip(illuminated, 0, 255).astype(np.uint8)
    illuminated_images.append(illuminated)

# Visualize
fig, axes = plt.subplots(1, 5, figsize=(16, 4))

for idx, (img, level) in enumerate(zip(illuminated_images, illumination_levels)):
    axes[idx].imshow(img)
    axes[idx].set_title(f'Illumination: {level:.2f}x')
    axes[idx].axis('off')

plt.tight_layout()
plt.show()

# Show pixel value changes
print("\nIllumination Effect on Pixel Values:")
test_y, test_x = h//2, w//2
original_pixel = original[test_y, test_x]

print(f"\nOriginal pixel at ({test_y}, {test_x}): RGB={original_pixel}")
print(f"\n{'Illumination':<15} {'R':<8} {'G':<8} {'B':<8} {'Gray (standard)':<15}")
print("-" * 55)

for level in illumination_levels:
    new_pixel = np.clip(original_pixel.astype(np.float32) * level, 0, 255).astype(np.uint8)
    gray = 0.299*new_pixel[0] + 0.587*new_pixel[1] + 0.114*new_pixel[2]
    print(f"{level:.2f}x{' '*9} {new_pixel[0]:<8} {new_pixel[1]:<8} {new_pixel[2]:<8} {gray:<15.2f}")
```

```python
# Summary and key insights
print("\n" + "=" * 80)
print("SUMMARY: KEY CONCEPTS IN PIXEL VALUES AND GRAYSCALE CONVERSION")
print("=" * 80)

summary = """
1. GRAYSCALE CONVERSION FORMULA:
   - Standard ITU-R BT.601: Gray = 0.299×R + 0.587×G + 0.114×B
   - Green has highest weight (0.587) because human eyes are most sensitive to green
   - Blue has lowest weight (0.114) because human eyes are least sensitive to blue

2. COLOR SPACES:
   - RGB: Hardware-native, used by cameras and displays
   - HSV: Perceptually intuitive, useful for color-based detection
   - YCbCr: Separates luminance from chrominance, used in video compression
   - LAB: Perceptually uniform, useful for color analysis
   - Grayscale: Single intensity channel, reduced storage and faster processing

3. ILLUMINATION-REFLECTANCE MODEL:
   - Pixel Intensity = Illumination × Reflectance
   - I(x,y) = L(x,y) × R(x,y)
   - Brightness depends on both light and material properties

4. PIXEL CALCULATIONS:
   - Always clip values to valid range [0, 255] after arithmetic operations
   - Use appropriate data types: uint8 (0-255), float32 (0.0-1.0)
   - Different conversion methods produce slightly different results

5. PRACTICAL CONSIDERATIONS:
   - Choose color space based on application needs
   - HSV is better for color-based object detection
   - Grayscale is better for simplification and edge detection
   - Use standard ITU-R formula for perceptually accurate grayscale
"""

print(summary)
```
