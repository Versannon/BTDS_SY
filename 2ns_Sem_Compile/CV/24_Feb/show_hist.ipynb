Displaying histogram of image wihtout using inbuilt function

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Display histogram of image "image.jpeg"

```python
img = cv2.imread("image.jpeg", 0)
```

```python
#histogram of image

height, width = img.shape
freq = np.zeros(256)

for i in range(height):
    for j in range(width):
        intensity = img[i][j]
        freq[intensity] += 1

#plotting histogram

plt.bar(range(256), freq)
plt.title("Histogram of image")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()
```

Show histogram Using inbuilt functions

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.plot(hist)
plt.title("Histogram using inbuilt Functions")
plt.show()
```
