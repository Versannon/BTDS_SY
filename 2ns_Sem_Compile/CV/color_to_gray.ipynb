Converting color to grayscale image without using inbuilt function

Using image "image.jpeg"

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpeg")

height, width, channels = img.shape[:3]
freq = np.zeros(256)
```

```python
Gray = img

#create empty grayscale image

gray = np.zeros((height, width), dtype = np.uint8)

#convert manually using for loop

for i in range(height):
    for j in range(width):
        B = img[i, j, 0]
        G = img[i, j, 1]
        R = img[i, j, 2]
        
        gray_value = int(0.299 * R + 0.587 * G + 0.114 * B)
        gray[i, j] = gray_value

```
