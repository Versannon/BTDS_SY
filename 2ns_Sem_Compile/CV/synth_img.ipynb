```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix = np.random.randint(0,256,(900,900))
print(matrix,type(matrix),"\n")
```

```python
img = np.array(matrix, dtype=np.uint8)
print(img,type(img),"\n")
```

```python
cv2.imwrite("sythn_img.jpg", img)
```

```python
plt.imshow(img, cmap='gray')
plt.title('Synthetic Grayscale Image')
plt.show()
```

```python
cv2.imshow("Synthetic Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Creating negation of the image
neg_img = 255 - img
cv2.imwrite("neg_sythn_img.jpg", neg_img)
cv2.imshow("neg_image", neg_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
