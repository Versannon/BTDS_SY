## 3: Histogram of image (using cv2's inbuilt function)

## 4: Histogram of image (without using inbuilt function)

```python
#histogram of image

height, width = img.shape[:2]
freq = np.zeros(256)

for i in range(height):
    for j in range(width):
        intensity = img[i][j]
        freq[intensity] += 1

#plotting histogram

plt.bar(range(256), freq)
plt.title("Histogram of image")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()
```

## 5: RGB to GRAY scale (without using inbuilt function)

```python
height, width, channels = img.shape[:3]
freq = np.zeros(256)

Gray = img

#create empty grayscale image

gray = np.zeros((height, width), dtype = np.uint8)

#convert manually using for loop

for i in range(height):
    for j in range(width):
        B = img[i, j, 0]
        G = img[i, j, 1]
        R = img[i, j, 2]
        
        gray_value = int(0.299 * R + 0.587 * G + 0.114 * B)
        gray[i, j] = gray_value

cv2.imshow('Color to gray', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 6: Thresholding (Inbuilt)

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

cv2.imshow("Thressholded Image", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 7: Thresholding (Without using inbuilt)

```python
# Manual threshold in the same loop style
thresh_manual = np.zeros((height, width), dtype=np.uint8)

for i in range(height):
    for j in range(width):
        if gray[i, j] >= 127:
            thresh_manual[i, j] = 255
        else:
            thresh_manual[i, j] = 0

thresh = cv2.cvtColor(thresh_manual, cv2.COLOR_GRAY2BGR)

cv2.imshow('Thresholding (Manual) | Left: Input Gray, Right: Output', thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 11: Contrast Stretching (Without using inbuilt)

```python
r_min = np.min(img)
r_max = np.max(img)

# Create output image
stretched = np.zeros(img.shape, dtype=np.uint8)

# Apply contrast stretching formula
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        stretched[i, j] = (img[i, j] - r_min) * 255 / (r_max - r_min)

# Display results

cv2.imshow('Irginal Image', img)
cv2.imshow('stretched image', stretched)
cv2.waitKey(0)
cv2.destroyAllWindows()


"""
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Contrast Stretched")
plt.imshow(stretched, cmap='gray')
plt.axis('off')

plt.show()"""
```

## 12: Histogram Equalization (using inbuilt function)

## 13: Histogram Equalization (Without using inbuilt)

```python
# Step 1: Calculate histogram
hist = np.zeros(256)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        hist[img[i, j]] += 1

# Step 2: Normalize histogram (PDF)
pdf = hist / (img.shape[0] * img.shape[1])

# Step 3: Compute cumulative distribution function (CDF)
cdf = np.cumsum(pdf)

# Step 4: Create transformation function
transform = np.round(cdf * 255)

# Step 5: Apply transformation
equalized = np.zeros(img.shape, dtype=np.uint8)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        equalized[i, j] = transform[img[i, j]]

cv2.imshow("Original Image", img)
cv2.imshow("Histogram Equalized Image", equalized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 14: Explanation (How the 6 techniques work)

1. **Histogram**: 
   - Counts the frequency of each grayscale intensity value from 0 to 255 in an image.
   - Visualized as a graph where the x-axis represents intensity levels and y-axis represents pixel count.
   - Useful for understanding the distribution of brightness in an image and identifying exposure issues.

2. **RGB to Gray**: 
   - Converts a color image (with separate R, G, B channels) into a grayscale image (single channel).
   - Uses weighted contributions: typically `Gray = 0.299*R + 0.587*G + 0.114*B` (based on human perception of colors).
   - Green has higher weight because human eyes are more sensitive to green light.
   - Reduces memory requirements and simplifies image processing.

3. **Thresholding**: 
   - Compares each pixel intensity to a fixed threshold value (e.g., 127).
   - Converts grayscale image to binary image: pixels below threshold become 0 (black), above become 255 (white).
   - Useful for segmentation and creating masks for object detection.

4. **Negation**: 
   - Inverts pixel values using the formula: `Inverted_Pixel = 255 - Original_Pixel`.
   - Creates a photographic negative effect where bright areas become dark and vice versa.
   - Useful for enhancing certain features or creating artistic effects.

5. **Contrast Stretching**: 
   - Expands the dynamic range of pixel values from their original range `[r_min, r_max]` to the full grayscale range `[0, 255]`.
   - Formula: `Stretched_Pixel = (Pixel - r_min) * 255 / (r_max - r_min)`.
   - Makes use of the entire available intensity range, improving visibility of details in dark or underexposed areas.

6. **Histogram Equalization**: 
   - Redistributes pixel intensity values using histogram CDF (Cumulative Distribution Function) mapping.
   - Aims to increase contrast by spreading out the most frequent intensity values.
   - Results in a more uniform histogram distribution, making image details more visible.
   - Particularly effective for underexposed or low-contrast images.
