Spatial Filtering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python

img_mat = img
print("\n Image Matrix:", img_mat)
```

```python
#initialize mask
kernel = np.array([[1,1,1],[1,1,1],[1,1,1]])/9
print("\n Kernel:", kernel)
```

```python
#prevent size reduction using padding
img_pad = np.pad(img_mat, pad_width = 1, mode ='edge')
```

```python
#find local region
local_reg = img_pad[0:3, 0:3]
print("\n Local Region:", local_reg)
```

```python
#apply mask to local region
filtered_reg = local_reg * kernel
print("\n Filtered Region:", filtered_reg)
```

```python
#new intensity value
new_intensity = np.ceil(np.sum(filtered_reg))
print("\n New Intensity Value:", new_intensity)
```

```python
def Myfilter2D(img, kernel):
    "Creating function work on gray scale img only"
    img_pad = np.pad(img, pad_width = 1, mode='reflect')
    M,N = img.shape
    Filtered_img = np.zeros((M,N), np.unit8)
    for i in range(M):
        for j in range(1, N+1):
            region = img_pad[i-1:i+2, j-1:j+2]
            Filtered_img[i-1, j-1] = np.round(np.sum(region * kernel))
    return Filtered_img
```
