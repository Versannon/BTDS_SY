```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#Load Image
img= mpimg.imread("sample.jpeg")        #RGB
img2= cv2.imread("sample.jpeg")        #BGR
```

```python
M,N,C=img.shape

#To calculate the number of pixels in the image.
pixels=M*N
print(pixels)
```

```python
#Display Image
plt.figure()

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Read Image using Matplotlib")

plt.subplot(1,2,2)
plt.imshow(img2)
plt.title("Read Image using OpenCv")

plt.show()
```

```python
# Convert the BGR image into RGB
img_rgb=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

plt.figure()

plt.subplot(1,3,1)
plt.imshow(img2)
plt.title("Read Image using OpenCv")

plt.subplot(1,3,2)
plt.imshow(img_rgb)
plt.title("BGR to RGB")

plt.subplot(1,3,3)
plt.imshow(img_gray,cmap="gray")
plt.title("RGB to GRAY")

plt.show()
```

```python
plt.figure()

plt.subplot(1,2,1)
plt.imshow(img_gray,cmap="gray")
plt.title("Gray Image")

plt.subplot(1,2,2)
plt.hist(img_gray.ravel(),bins=256,range=[0,256])
plt.title("Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
```
